```python
import requests
BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel_advisor_agent"
USER_ID = "user_123"
SESSION_ID = "3acfcdb5-093a-4a44-8cf1-ff3de1cb31de"
```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")
print(response.json())
```

```python
message = {
    "appName": "APP_NAME",
    "userId": "USER_ID",
    "sessionId": "SESSION_ID",
    "newMessage":{
        "parts":[
            {"text":"I want to go to Tokyo"}
        ],
        "role":"user"
    }
}
response = requests.post(f"{BASE_URL}/run", json=message)

print(response.json())
```

```python
data= response.json()
for event in data:
    content = event.get("content", "")
    parts = content.get("parts", [])
    for part in parts:
        function_call = part.get("function_call", None)
        if function_call:
            print(function_call.get("name", ""))
        text = part.get("text", None)
        if text:
            print(text)
    print("="*60)
```
